# 01 — Train & Evaluate (Frozen CLIP + Linear Head)
Set `DATA_ROOT` to your Egyptian dataset folder (folder-per-class). Then run all cells.


```python
import sys
from pathlib import Path

ROOT = Path("..").resolve()  
sys.path.insert(0, str(ROOT))

from src.config import TrainConfig
from src.utils import seed_everything

DATA_ROOT = str((ROOT / "data" / "Egyptian").resolve())
ARTIFACTS_DIR = str((ROOT / "artifacts").resolve())

cfg = TrainConfig(
    data_root=DATA_ROOT,
    artifacts_dir=ARTIFACTS_DIR,
    device="cuda",
    model_name="ViT-B-32-quickgelu",
    pretrained="openai",
    batch_size=64,
    epochs=10,
    lr=1e-3,
    num_workers=4,
    use_amp=True,
)

seed_everything(cfg.seed)
cfg
```

```python
from src.train import train_linear_head

ckpt_path, train_metrics = train_linear_head(cfg)
ckpt_path, train_metrics
```

```python
from src.eval import eval_linear_ckpt, eval_zeroshot, save_metrics

metrics_linear = eval_linear_ckpt(
    data_root=cfg.data_root,
    ckpt_path=ckpt_path,
    artifacts_dir=cfg.artifacts_dir,
    device=cfg.device,
    batch_size=cfg.batch_size,
    num_workers=cfg.num_workers,
)
metrics_linear
```

```python
metrics_zs = eval_zeroshot(
    data_root=cfg.data_root,
    artifacts_dir=cfg.artifacts_dir,
    device=cfg.device,
    model_name=cfg.model_name,
    pretrained=cfg.pretrained,
    batch_size=cfg.batch_size,
    num_workers=cfg.num_workers,
)
metrics_zs
```

```python
metrics = {}
metrics.update(metrics_linear)
metrics.update(metrics_zs)

out_path = str(Path(cfg.artifacts_dir) / "reports" / "metrics.json")
save_metrics(metrics, out_path)
out_path, metrics
```
